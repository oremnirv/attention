```python
from data import data_generation, data_picker, batch_creator, gp_kernels
from helpers import metrics
import sklearn.gaussian_process as gp 
import matplotlib.pyplot as plt
import numpy as np
import scipy
plt.style.use('ggplot')

import warnings
warnings.filterwarnings("ignore")
import pandas as pd
%matplotlib inline
```

```python
def experiment(extrapo):
    if extrapo:
        x = np.load('/Users/omernivron/Downloads/GPT_data_goldstandard/x_extra.npy')
        y = np.load('/Users/omernivron/Downloads/GPT_data_goldstandard/y_extra.npy')
    else:
        x = np.load('/Users/omernivron/Downloads/GPT_data_goldstandard/x_interpol.npy')
        y = np.load('/Users/omernivron/Downloads/GPT_data_goldstandard/y_interpol.npy')
    return x, y
```

```python
def predict_golden(t, x, x_star):
    n = x.shape[0]
    xx = np.concatenate((x, x_star), axis = 0)
    d = np.zeros((n, n))
    np.fill_diagonal(d, 0.00001)
    K_inv = np.linalg.inv(gp_kernels.rbf_kernel(x) + d) 
    k_t = (gp_kernels.rbf_kernel(xx.reshape(-1, 1))[:-1, -1]).transpose() 
    μ = np.dot(np.dot(k_t, K_inv), t)
    σ_sq = 1 - np.dot(np.dot(k_t, K_inv), k_t.transpose())
    return μ, σ_sq
```

```python
# GP independent sequence learning
m1 = []; m2 = []; r = []; r2 = []; ku = [];  ku2= []
for boolean in [True, False]:
    x, y = experiment(extrapo = boolean)
    online = False; mse_metric = 0; r_sq_metric = 0; 
    kuee_metric = 0; kuee_metric_g = 0
    μ = []; σ = []; k = []; ρ= [];
    μ_g = []; σ_g = [];
    m = int(x.shape[0] / 10)
    y_mean = np.mean(y[:m, :40])
    y_te = y[:m, 40]
    for j in range(0, m):
        λ=np.exp(-1); ν = 1; σ1 = 0.0001;
        kernel = ν**2 * gp.kernels.RBF(length_scale=λ)
        model = gp.GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer = 3, normalize_y=True)
        x_tr =  x[j, :40].reshape(-1, 1); y_tr = y[j, :40].reshape(-1, 1)

        model.fit(x_tr, y_tr)
        μ_te, σ_te = model.predict(x[j, 40].reshape(-1, 1), return_std=True)
        μ_gold, σ_sq_gold = predict_golden(y_tr, x_tr, x[j, 40].reshape(-1, 1))
        μ.append(μ_te[0][0]); σ.append(σ_te[0])
        μ_g.append(μ_gold[0]); σ_g.append(σ_sq_gold)
        k.append(model.kernel_.get_params()['k2'])
        ρ.append(model.kernel_.get_params()['k1'])

        if (j % 400 == 0):
            print('J: ', j)
            plt.subplot(1, 2, 1)
            axes = plt.gca()
            axes.set_ylim([-2, 2])
            plt.scatter(x_tr, y_tr, c = 'black')
            plt.scatter(x[j, 40], y_te[j], c='purple', s = 24)
            plt.scatter(x[j, 40], μ_gold, c='red')
            plt.errorbar(x = x[j, 40], y = (μ_gold[0]), yerr = 2 * np.sqrt(σ_sq_gold), fmt='o', ecolor='g', capthick=2)
            plt.title('Gold standard')


            plt.subplot(1, 2, 2)
            axes = plt.gca()
            axes.set_ylim([-2, 2])
            plt.scatter(x_tr, y_tr, c = 'black')
            plt.scatter(x[j, 40], y_te[j], c='purple', s = 24)
            plt.scatter(x[j, 40], μ_te[0], c='red')
            plt.errorbar(x = x[j, 40], y = (μ_te[0]), yerr = 2 * (σ_te), fmt='o', ecolor='g', capthick=2)
            plt.title('Predicted gold standard')
            plt.show()
            
            print('y_te: ', y_te[j])
            print('μ_te: ', μ_te[0][0])
            print('kuee_interval: [{}, {}]'.format(μ_te[0] - 2 * σ_te, μ_te[0] + 2 * σ_te)) 
            print('METRIC KUEE: ', metrics.KUEE(y_te[j], μ_te[0], σ_te))

        kuee_metric += metrics.KUEE(y_te[j], μ_te[0], σ_te)
        kuee_metric_g += metrics.KUEE(y_te[j], μ_gold[0], σ_sq_gold)



    mse_metric = metrics.mse(y_te, μ)
    mse_metric_g = metrics.mse(y_te, μ_g)


    r_sq_metric = metrics.r_squared(y_te, μ, y_mean)
    r_sq_metric_g = metrics.r_squared(y_te, μ_g, y_mean)

    mse_metric *= (1 / m); mse_metric_g *= (1 / m)
    kuee_metric *= (1 / m); kuee_metric_g *= (1 / m)
    m1.append(mse_metric); m2.append(mse_metric_g)
    ku.append(kuee_metric); ku2.append(kuee_metric_g)
    r.append(r_sq_metric); r2.append(r_sq_metric_g)
    
summary={'MSE': np.array(m1), 
           'MSE_gold': np.array(m2),
           'KUEE': np.array(ku),
           'KUEE_gold': np.array(ku2), 
           'R': np.array(r), 
           'R_gold': np.array(r2)}
```
